```python
import pandas as pd
```

```python
df = pd.read_csv("../data/California_Houses.csv")
df.fillna(0, inplace=True)
df
```

```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "notebook_connected"
```

```python
fig1 = px.scatter(df, x="Distance_to_coast", y="Median_House_Value", color="Households", size="Median_Income")
fig1.update_layout(title="California Housing Prices")
fig1.show()
```

```python
fig2 = px.scatter_3d(df, x="Distance_to_coast", y="Median_House_Value", z="Population", color="Households", size="Median_Income")
fig2.update_layout(title="3D Plot")
fig2.show()
```

```python
df_scaled = df[["Median_House_Value", "Median_Income", "Median_Age", "Tot_Rooms", "Tot_Bedrooms", "Population", "Households", "Distance_to_coast"]].copy()

for col in df_scaled.columns:
    df_scaled[col] = (df_scaled[col] - df_scaled[col].min()) / (df_scaled[col].max() - df_scaled[col].min())

df_scaled
```
